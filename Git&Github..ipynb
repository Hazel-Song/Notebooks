# 3.Git Commond
📨目录：
1. 新建本地存储库`git init`
2. 将本地git链接到远程储存库`git remote add`
3. 将分支重命名为`main`
4. 暂存并提交更改`git add .` `git commit`
5. 检查文件状态`git status`
6. 将代码推送到远程储存库`git push`
   


## 3.1 Git初始化 `git init`
`git init` 会初始化一个新的Git仓库，运行此命令时，Git 将创建一个.git以你的项目文件夹命名的新目录,此目录包含有关项目历史记录和配置的所有信息。成功完成后，终端将返回以下内容：
`Initialized empty Git repository in /Users/username/.git`

## 3.2 Git远程创建 `git remote add origin <url>`
我们需要将本地仓库（计算机上的文件夹）与远程仓库（GitHub 上的链接）连接起来。
我们可以使用以下remote add命令：
`git remote add origin repository_url`

## 3.3 Git分支 `git branch -M main`
创建一个默认分支main：
`git branch -M main`
- 打开代码编辑器创建一个终端（zsh是mac默认终端名称），使用`pwd`指令确认仓库位置，应该会收到包含当前目录路径的响应！
- 要检查我们是否已成功连接本地和远程存储库并重命名分支，请运行以下命令：`git branch
// Output:
// * main`如果你在输出中看到main，就说明一切就绪！🎉

## 3.4 Git Staging area/Index `git add`和`git commit -m "xxx"`
我们前面已经连接了本地和远程存储库，让我们学习如何在它们之间移动代码。
**Working dictionary**是您在计算机上正在处理或更改的文件和文件夹所在的位置！当您对文件进行更改时，Git 将跟踪这些更改并允许您将它们移动到**Staging Area**。
**Staging Area**是项目内文件和文件夹的状态，它是一个临时区域，您可以在其中选择要“提交”或发送到远程存储库的文件。
使用`git add`命令就可以做到这一点！以下是一些您可能需要考虑的命令变体：
- `git add example.txt`会将单个文件添加到暂存区。如果您只想让某个工作文件显示在 GitHub 上，而不是项目的其余部分，请考虑指定一个特定的文件。
- `git add .`会将所有更改的文件添加到暂存区。这意味着任何在工作目录中添加或更改的新文件都将包含在内。
- `git add *.html`只会暂存指定扩展名的文件。在本例中，此命令只会添加.html文件。

**Commited file**就是将当前代码以“快照”的形式共享给 GitHub 在线用户。提交帮助我们区分代码中的不同操作，并允许我们添加有用的信息，以便了解和跟踪代码变更。📸
`git commit -m 'Your commit message here!'`

返回终端或命令提示符，并导航到项目文件夹。您将使用git add .将项目文件夹中的所有文件提交到暂存区。然后，使用git commit -m 'Initial Commit'编写您的第一条提交消息！

## 3.5 Git提交 `git push -u origin main`
用于将本地的更改提交到远程储存库，一般使用`git push`，若第一次提交，通常使用`git push -u origin main`设置默认提交分支并将本地分支与远程分支链接起来，以后都将默认使用该分支。

## 3.6 Git状态
`git status`用来查看：
- 暂存文件已准备好提交。
- 未暂存的文件尚未准备好提交。
- 未跟踪的文件是 Git 以前未见过的新文件。

# 4.Git co-work
## 克隆
拉取他人的在线仓库内容：`git clone https://github.com/github-user/repo-name.git` ，
或使用**fork**功能获得分支副本到个人空间进行修改

## 在分支上工作

我们使用以下命令来实现：
- 从你当前所在的分支创建一个新分支：`git branch new-branch-name`
- 切换到该分支：`git checkout new-branch-name`
- 通过单个命令创建并切换到新分支：`git checkout -b <new-branch-name> `
